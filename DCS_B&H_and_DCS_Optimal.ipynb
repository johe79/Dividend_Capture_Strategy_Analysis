```python
# Import libraries

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import csv
import yfinance as yf
from datetime import timedelta
import statsmodels.api as sm
import matplotlib.dates as mdates
```

```python
#

gww = pd.read_csv("gww.csv")
wst = pd.read_csv("wst.csv")
apd = pd.read_csv("apd.csv")
low = pd.read_csv("low.csv")

gww['symbol'] = gww['symbol'].replace("DA Selected - DCS Optimal", "GWW")
wst['symbol'] = wst['symbol'].replace("DA Selected - DCS Optimal", "WST")
apd['symbol'] = apd['symbol'].replace("DA Selected - DCS Optimal", "APD")
low['symbol'] = low['symbol'].replace("DA Selected - DCS Optimal", "LOW")


print(gww)
print(wst)
print(apd)
print(low)
```

```python
#

consolidated = pd.concat([gww,wst,apd,low], axis = 0)
#print(consolidated)

consolidated = consolidated.groupby(["year_month"]).agg(percentage_gains = ("percentage_gains","mean"))

consolidated["symbol"] = "DA Selected - B&H"

print(consolidated)

consolidated.to_csv("DA_Selected_B&H.csv")
```

```python
#

gww_optimal = pd.read_csv("gww_optimal.csv")
wst_optimal = pd.read_csv("wst_optimal.csv")
apd_optimal = pd.read_csv("apd_optimal.csv")
low_optimal = pd.read_csv("low_optimal.csv")

gww_optimal['symbol'] = gww_optimal['symbol'].replace("DA Selected - DCS Optimal", "GWW")
wst_optimal['symbol'] = wst_optimal['symbol'].replace("DA Selected - DCS Optimal", "WST")
apd_optimal['symbol'] = apd_optimal['symbol'].replace("DA Selected - DCS Optimal", "APD")
low_optimal['symbol'] = low_optimal['symbol'].replace("DA Selected - DCS Optimal", "LOW")

#print(gww_optimal)
#print(wst_optimal)
#print(apd_optimal)
#print(low_optimal)
```

```python
#

consolidated_optimal = pd.concat([gww_optimal,wst_optimal,apd_optimal,low_optimal], axis = 0)
#print(consolidated)

consolidated_optimal = consolidated_optimal.groupby(["year_month"]).agg(percentage_gains = (
    "percentage_gains", "mean"))

consolidated_optimal["symbol"] = "DA Selected - DCS Optimal"

print(consolidated_optimal)

consolidated_optimal.to_csv("DA_Selected_Optimal.csv")
```

```python
#

b_h = pd.read_csv("top_10_results.csv")
print(b_h)
```

```python
#

symbols = ['IVV','SDY','DA Selected - DCS Halfway']

b_h = b_h[b_h['symbol'].isin(symbols)]
print(b_h)
# Rename specific cells in the "symbol_start" column
b_h["symbol"] = b_h["symbol"].replace({"IVV": "SP 500 - B&H (IVV)", "SDY": "DA ALL - B&H (SDY)"})

columns = ["year_month", "percentage_gains", "symbol"]
b_h = b_h[columns] 


consolidated_optimal["year_month"] = consolidated_optimal.index
consolidated["year_month"] = consolidated.index

all_strategies = pd.concat([b_h,consolidated, consolidated_optimal], axis = 0).reset_index()
#print(all_strategies)

# Ensure 'year_month' is in datetime format
all_strategies['year_month'] = pd.to_datetime(all_strategies['year_month'])

plt.figure(figsize=(18, 10))
sns.lineplot(
    data=all_strategies, x='year_month', y='percentage_gains', hue='symbol', linewidth=2.5, palette='tab10')

# Annotate the last point of each line with the stock symbol
for symbol in all_strategies['symbol'].unique():
    symbol_data = all_strategies[all_strategies['symbol'] == symbol]
    last_date = symbol_data['year_month'].max()
    last_value = symbol_data[symbol_data['year_month'] == last_date]['percentage_gains'].values[0]
    plt.text(last_date, last_value, symbol, horizontalalignment='left', size='medium', color='black', weight='semibold')

# Formatting the x-axis to show only years
plt.gca().xaxis.set_major_locator(mdates.YearLocator())
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y'))

plt.title('Percentage Gains Over Time for Selected Stocks', fontsize=18)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Percentage Gains (%)', fontsize=14)
plt.legend(title='Stock Symbol', loc='upper left', bbox_to_anchor=(0.05, 0.95), fontsize=12)
plt.grid(True)
plt.tight_layout()
plt.show()
plt.close()
```
